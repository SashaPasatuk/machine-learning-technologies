## 1. Текстовое описание набора данных
Набор данных о подержанных автомобилях Toyota в Великобритании. Набор данных содержит следующие колонки:
- model - модель автомобиля
- year - год выпуска
- price - цена (в фунтах стерлингах)
- transmission - тип коробки передач
- mileage - пробег (в милях)
- fuelType - тип используемого топлива
- tax - налог на автомобиль
- mpg - расход топлива (миль на галон)
- engineSize - объем двигателя
